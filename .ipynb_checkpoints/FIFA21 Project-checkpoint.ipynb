```python
#Changign BP values to strings
#We are also going to merge some of the columns to reduce the amount variance in this column
#To do this we are going to use a function and mapping
#We will be merging similar positions

def change_position(pos):
    if(pos == "CF"):
        return "ST"
    elif(pos == "RWB"):
        return "RB"
    if(pos == "LWB"):
        return "LW"
    elif(pos == "CDM"):
        return "CM"  
    return pos
        
data['bp'] = list(map(change_position, data['bp']))
```

```python
data['bp'].value_counts()
```

```python
#Now we are going to remove the "lbs" from the weight column and the " " " and the " ' " from the height column.
#To do this we are going to use a regex (regular expression) that extracts digits.

data['height'].value_counts()
```

```python
# Using regex in our function, we will pass a number to it
#- If the outcome from re.findall (finds every occurence of a specified regex) has more than 1 outcome, it will be because
#  of our height column which has 2 different numbers. One for feet and one for inches. I will then convert that output into 
#  inches.
#- If there is 1 outcome it will most likely be our weight.
# With the following logic from above we will format the return statements to match what we want so we can reuse the formula.

def set_ints(string_input):
    output = re.findall("\d+", string_input)
    if len(output) == 1:
        return int(output[0])
    else:
        return (int(output[0]) * 12 + int(output[1]))

data['height'] = list(map(set_ints, data['height']))
data['height'].value_counts()
```

```python
#Checking values for weight now.
data['weight'].value_counts()
```

```python
#Reusing the formula that we made for height we will now pass weight into it.
data['weight'] = list(map(set_ints, data['weight']))
data['weight'].value_counts()
```

```python
#Now that we have the value we need in the correct format, we will convert the dtype for each column.
#First we're gonna double check the dtypes
data.dtypes
```

```python
#The dtypes are automatically converted to int because our formula returns int casted numbers
data['weight'] = data['weight'].astype("int64")
data['height'] = data['height'].astype("int64")
```

```python
#We will now seperate the data based on y,x,numerical, and categorical
Y = data['ova']
data = data.drop(['ova'], axis=1)
X_num = data.select_dtypes(include = np.number)
X_cat = data.select_dtypes(include = np.object)
```

```python
#We will start transofrming and encoding our data for the purpose of setting up for our linear regression models
#Im going to use 3 different training/testing sizes.
transformer = Normalizer().fit(X_num)
x_normalized = transformer.transform(X_num)
encoder = OneHotEncoder().fit(X_cat)
encoded = encoder.transform(X_cat).toarray()
X = np.concatenate((x_normalized, encoded), axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
predictions  = lm.predict(X_test)
print(r2_score(y_test, predictions), mean_squared_error(y_test, predictions))
```
